# Generative Adversarial Networks (GANs) - Q&A

## Q1: Explain the minimax loss function in GANs and how it ensures competitive training between the generator and discriminator.

GANs are modeled as a two-player minimax game where the generator \( G \) tries to minimize the ability of the discriminator \( D \) to distinguish between real and fake data, while the discriminator tries to maximize its accuracy. The objective function is:

$$
    \min_G \max_D V(D, G) = \mathbb{E}_{x \sim p_{\text{data}}(x)} [\log D(x)] + \mathbb{E}_{z \sim p_z(z)} [1 - \log D(G(z))]
$$

Here, \( x \) is a real data sample, \( z \) is random noise, and \( G(z) \) is the generated data. This loss function ensures that both networks continuously improve—\( D \) becomes better at classification, and \( G \) learns to generate more realistic samples that can fool \( D \).

---

## Q2: What is mode collapse, why can it occur during GAN training, and how can it be mitigated?

Mode collapse happens when the generator produces only a limited variety of outputs, generating the same or similar samples repeatedly instead of a diverse set of realistic outputs. This occurs because the generator finds a particular output that fools the discriminator consistently and keeps replicating it rather than exploring the full data distribution.

**Mitigation Strategies:**
- **Batch Normalization**: Helps stabilize training.
- **Minibatch Discrimination**: Encourages diversity in generated samples.
- **Wasserstein GAN (WGAN)**: Uses the Wasserstein distance metric for a more stable loss function.

---

## Q3: Explain the role of the discriminator in adversarial training.

The discriminator \( D \) is responsible for distinguishing between real data from the dataset and fake data generated by the generator \( G \). It serves as the adversary in the training process by providing feedback to \( G \), which helps improve the quality of generated samples. As training progresses, \( D \) continuously refines its ability to detect fake samples, forcing \( G \) to produce increasingly realistic outputs.

---

## Q4: How do metrics like IS and FID evaluate GAN performance?

Although the document does not explicitly mention **Inception Score (IS)** and **Fréchet Inception Distance (FID)**, here’s how they typically evaluate GAN performance:

- **Inception Score (IS)**: Measures the quality and diversity of generated images by evaluating the entropy of class probabilities assigned by a pretrained classifier (e.g., Inception network). High IS suggests more diverse and realistic images.
- **Fréchet Inception Distance (FID)**: Measures the distance between the real and generated data distributions using feature representations from a deep neural network. Lower FID indicates better GAN performance as it suggests a smaller difference between real and generated data distributions.


## MNIST DATASET

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import os
from tensorflow.keras import layers

# Load and preprocess data
(train_images, _), (_, _) = tf.keras.datasets.mnist.load_data()
train_images = train_images.reshape(-1, 28, 28, 1).astype('float32')
train_images = (train_images - 127.5) / 127.5  # Normalize to [-1, 1]

BUFFER_SIZE = 60000
BATCH_SIZE = 256
train_dataset = tf.data.Dataset.from_tensor_slices(train_images).shuffle(BUFFER_SIZE).batch(BATCH_SIZE)

# Create directory for saving images
os.makedirs("generated_images", exist_ok=True)

def make_generator_model():
    model = tf.keras.Sequential([
        layers.Dense(7*7*256, use_bias=False, input_shape=(100,)),
        layers.BatchNormalization(),
        layers.LeakyReLU(),
        layers.Reshape((7, 7, 256)),
        layers.Conv2DTranspose(128, (5, 5), strides=(1, 1), padding='same', use_bias=False),
        layers.BatchNormalization(),
        layers.LeakyReLU(),
        layers.Conv2DTranspose(64, (5, 5), strides=(2, 2), padding='same', use_bias=False),
        layers.BatchNormalization(),
        layers.LeakyReLU(),
        layers.Conv2DTranspose(1, (5, 5), strides=(2, 2), padding='same', use_bias=False, activation='tanh')
    ])
    return model

def make_discriminator_model():
    model = tf.keras.Sequential([
        layers.Conv2D(64, (5, 5), strides=(2, 2), padding='same', input_shape=[28, 28, 1]),
        layers.LeakyReLU(),
        layers.Dropout(0.3),
        layers.Conv2D(128, (5, 5), strides=(2, 2), padding='same'),
        layers.LeakyReLU(),
        layers.Dropout(0.3),
        layers.Flatten(),
        layers.Dense(1)
    ])
    return model

cross_entropy = tf.keras.losses.BinaryCrossentropy(from_logits=True)

def discriminator_loss(real_output, fake_output):
    real_loss = cross_entropy(tf.ones_like(real_output), real_output)
    fake_loss = cross_entropy(tf.zeros_like(fake_output), fake_output)
    return real_loss + fake_loss

def generator_loss(fake_output):
    return cross_entropy(tf.ones_like(fake_output), fake_output)

generator = make_generator_model()
discriminator = make_discriminator_model()

generator_optimizer = tf.keras.optimizers.Adam(1e-4)
discriminator_optimizer = tf.keras.optimizers.Adam(1e-4)

def generate_and_save_images(model, epoch, test_input):
    predictions = model(test_input, training=False)
    fig = plt.figure(figsize=(4, 4))
    for i in range(predictions.shape[0]):
        plt.subplot(4, 4, i+1)
        plt.imshow(predictions[i, :, :, 0] * 127.5 + 127.5, cmap='gray')
        plt.axis('off')
    plt.savefig(f'generated_images/image_at_epoch_{epoch}.png')
    plt.show()

seed = tf.random.normal([16, 100])  # Fixed seed for consistent visualization

@tf.function
def train_step(images):
    noise = tf.random.normal([BATCH_SIZE, 100])
    with tf.GradientTape() as gen_tape, tf.GradientTape() as disc_tape:
        generated_images = generator(noise, training=True)
        real_output = discriminator(images, training=True)
        fake_output = discriminator(generated_images, training=True)
        gen_loss = generator_loss(fake_output)
        disc_loss = discriminator_loss(real_output, fake_output)
    gradients_of_generator = gen_tape.gradient(gen_loss, generator.trainable_variables)
    gradients_of_discriminator = disc_tape.gradient(disc_loss, discriminator.trainable_variables)
    generator_optimizer.apply_gradients(zip(gradients_of_generator, generator.trainable_variables))
    discriminator_optimizer.apply_gradients(zip(gradients_of_discriminator, discriminator.trainable_variables))
    return gen_loss, disc_loss

def train(dataset, epochs):
    for epoch in range(1, epochs + 1):
        gen_loss_epoch = 0
        disc_loss_epoch = 0
        num_batches = 0
        for image_batch in dataset:
            gen_loss, disc_loss = train_step(image_batch)
            gen_loss_epoch += gen_loss
            disc_loss_epoch += disc_loss
            num_batches += 1
        
        # Average loss over batches
        gen_loss_epoch /= num_batches
        disc_loss_epoch /= num_batches
        
        print(f'Epoch {epoch}/{epochs}, Generator Loss: {gen_loss_epoch:.4f}, Discriminator Loss: {disc_loss_epoch:.4f}')
        
        if epoch % 10 == 0:
            generate_and_save_images(generator, epoch, seed)

# Run training
train(train_dataset, epochs=50)
```
